```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('data/multiTimeline.csv', skiprows=1)
df.head()
```

```python
df.columns = (['month','mask1','skinCare1','mask','skinCare'])
df.head()
```

```python
df.month = pd.to_datetime(df.month)
df.set_index('month', inplace=True)
```

```python
df.head()
```

```python
df.plot(figsize=(20,10), linewidth = 5, fontsize=22)
plt.xlabel('Year', fontsize= 22);
```

```python
df.drop(['mask1', 'skinCare'],axis=1, inplace=True);
```

```python
df.plot(figsize=(20,10), linewidth = 5, fontsize=22)
plt.xlabel('Year', fontsize= 22);
```

```python
mask = df[['mask']]
mask.rolling(12).mean().plot(figsize=(20,10), linewidth = 5, fontsize=22)
plt.xlabel('Year', fontsize= 22);
```

```python
skinCare1 = df[['skinCare1']]
skinCare1.rolling(12).mean().plot(figsize=(20,10), linewidth = 5, fontsize=22)
plt.xlabel('Year', fontsize= 22);
```

```python
df_rm = pd.concat([mask.rolling(12).mean(),skinCare1.rolling(12).mean()], axis=1)
df_rm.plot(figsize=(20,10), linewidth = 5, fontsize=22)
plt.xlabel('Year', fontsize= 22);
```
